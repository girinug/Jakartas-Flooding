## Clustering

```python
new = file.filter(['jumlah_terdampak_rt','ketinggian_air_rata','jumlah_terdampak_jiwa'], axis=1)
new_data = new.copy()
```

```python
new.head()
```

```python
from sklearn import preprocessing
from sklearn.decomposition import PCA
X_scaled = preprocessing.scale(new)
pca = PCA(n_components=2)
df = pca.fit_transform(X_scaled)
df = pd.DataFrame(df)
```

```python
n_init = 10
max_iter = 300
tol = 0.0001
n_jobs = 2

plt.figure(figsize=(10, 8))
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42,n_init=n_init,
                   max_iter=max_iter,tol = tol,n_jobs=n_jobs)
    kmeans.fit(df)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
n_clusters = 4
kmeans = KMeans(n_clusters = n_clusters, init = 'k-means++', random_state = 42,n_init=n_init,
                   max_iter=max_iter,tol = tol,n_jobs=n_jobs)
kmeans.fit(df)
y_kmeans = kmeans.predict(df)
centers = kmeans.cluster_centers_
plt.scatter(df.loc[:, 0], df.loc[:, 1], c=y_kmeans, s=50, cmap='viridis')
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5)
plt.show()
```

```python
new['cluster'] =  pd.Series(y_kmeans)
```

```python
new = new[['jumlah_terdampak_rt','ketinggian_air_rata','jumlah_terdampak_jiwa','cluster']]
new.head()
```

```python
num = new_data.columns[:]

def cl(n):
    value = []
    for i in num:
        value.append(new.loc[new['cluster']==n,:].describe().loc['mean',i])
    return value
```

```python
num
```

```python
new.loc[new['cluster'] == 3, 'cluster'] = 4
new.loc[new['cluster'] == 2, 'cluster'] = 3
new.loc[new['cluster'] == 1, 'cluster'] = 2
new.loc[new['cluster'] == 0, 'cluster'] = 1
```

```python
cluster_1 = cl(1)
cluster_2 = cl(2)
cluster_3 = cl(3)
cluster_4 = cl(4)
```

```python
x = 0
w= 1
plt.figure(figsize=(15,9))
label = ['jumlah rt','ketinggian air','Jumlah warga terdampak']
for i in range(1,4):
    plt.subplot(2, 3, i)
    fig1 = plt.bar(x+1,cluster_1[i-1],color='y')
    fig2 = plt.bar(x+w+1,cluster_2[i-1],color='g')
    fig3 = plt.bar(x+3*w,cluster_3[i-1],color='r')
    fig4 = plt.bar(x+4*w,cluster_4[i-1],color='b')
    plt.title(label[i-1])
    plt.legend(['1','2','3','4'])
plt.savefig('graph1.png')
plt.show()
```

<p> Cluster 1 = Ringan. Terjadi pada area yang kecil (sekitar 5 RT), ketinggian air rata - rata berkisar 40 cm dan jumlah warga terdampak kurang dari 500 orang
<p> Cluster 2 = Sedang. Terjadi pada area yang kecil (sekitar 5 RT), tinggian air rata - rata mencapai 100 cm dan jumlah warga yang terdampak kurang dari 500 orang
<p> Cluster 3 = Sedang-Berat. Terjadi Pada area yang cukup luas yaitu lebih dari 10 RT dengan ketinggian air rata - rata mencapai 80 cm dengan jumlah rata - rata warga yang terdampak berkisar 1500 orang
<p> Cluster 4 = Berat. Terjadi pada area yang sangat luas dengan sekitar 30 RT terendam banjir dengan ketinggian air rata - rata lebih dari 80 cm dan jumlah warga terdampak mencapai 3000 orang
