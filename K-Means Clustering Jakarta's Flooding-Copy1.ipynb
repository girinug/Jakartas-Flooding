```python
file = file_awal.loc[file_awal['ketinggian_air_rata'] != 0]
```

## Clustering

```python
new = file.filter(['jumlah_terdampak_rt','ketinggian_air_rata','jumlah_terdampak_jiwa'], axis=1)
new_data = new.copy()
```

```python
new.head()
```

```python
from sklearn import preprocessing
from sklearn.decomposition import PCA
X_scaled = preprocessing.scale(new)
pca = PCA(n_components=2)
df = pca.fit_transform(X_scaled)
df = pd.DataFrame(df)
```

```python
n_init = 10
max_iter = 300
tol = 0.0001
n_jobs = 2

plt.figure(figsize=(10, 8))
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42,n_init=n_init,
                   max_iter=max_iter,tol = tol,n_jobs=n_jobs)
    kmeans.fit(df)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
n_clusters = 4
kmeans = KMeans(n_clusters = n_clusters, init = 'k-means++', random_state = 42,n_init=n_init,
                   max_iter=max_iter,tol = tol,n_jobs=n_jobs)
kmeans.fit(df)
y_kmeans = kmeans.predict(df)
centers = kmeans.cluster_centers_
plt.scatter(df.loc[:, 0], df.loc[:, 1], c=y_kmeans, s=50, cmap='viridis')
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5)
plt.show()
```

```python
new['cluster'] =  pd.Series(y_kmeans)
```

```python
new = new[['jumlah_terdampak_rt','ketinggian_air_rata','jumlah_terdampak_jiwa','cluster']]
new.head()
```

```python
num = new_data.columns[:]

def cl(n):
    value = []
    for i in num:
        value.append(new.loc[new['cluster']==n,:].describe().loc['mean',i])
    return value
```

```python
cluster_0 = cl(0)
cluster_1 = cl(1)
cluster_2 = cl(2)
cluster_3 = cl(3)
```

```python
num
```

```python
x = 0
w= 1
plt.figure(figsize=(15,9))
label = ['jumlah rt','ketinggian air','Jumlah warga terdampak']
for i in range(1,4):
    plt.subplot(2, 3, i)
    fig1 = plt.bar(x,cluster_0[i-1],color='y')
    fig2 = plt.bar(x+w,cluster_1[i-1],color='g')
    fig3 = plt.bar(x+2*w,cluster_2[i-1],color='r')
    fig4 = plt.bar(x+3*w,cluster_3[i-1],color='b')
    plt.title(label[i-1])
    plt.legend(['0','1','2','3','4'])
plt.savefig('graph1.png')
plt.show()
```

```python
new.loc[new['cluster'] == 3, 'cluster'] = 4
new.loc[new['cluster'] == 2, 'cluster'] = 3
new.loc[new['cluster'] == 1, 'cluster'] = 2
new.loc[new['cluster'] == 0, 'cluster'] = 1
```

```python
new = new.filter(['cluster'], axis=1)
data_baru = pd.concat([file_awal, new], axis=1)
data_baru.tail()
```

```python
data_baru['cluster'] = data_baru['cluster'].replace(np.nan, 0)
```
